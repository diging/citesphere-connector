```python
def get_file_ids(zotero_id,collection_id):
    testing = connector.get_collection_items(zotero_id,collection_id)
    file_ids=[]
    for x in testing["items"]:
        if type(x["gilesUploads"]) == type([1,2,3]) and len(x["gilesUploads"])>0:
            for y in x["gilesUploads"]:
                 file_ids.append(y["uploadedFile"]["id"]) 
```

```python
def get_all_collections(zotero_id):
    collections_ = connector.get_collections(zotero_id)
    collections_keys = [c["key"] for c in collections_["collections"]]
    return {zotero_id: collections_keys}
```

```python
def Merge(dict1, dict2):
    return(dict2.update(dict1))

collections_dict={}
for x in group_ids:
    Merge(get_all_collections(x),collections_dict)
collections_dict
```

```python
def get_items(group_id):
    #get items keys
    #{collections_id: [keys]}
    key_dict={}
    key_list=set()
    dev_dict=connector.get_collections(group_id)
    for z in dev_dict["collections"]:
        num_pages = math.ceil(z["numberOfItems"]/50)
        if num_pages == 1:
            items_dict= connector.get_collection_items(group_id, z["key"])
            for x in items_dict["items"]:
                key_list.add(x["key"])
        else:
            for i in range(num_pages):
                items_dict=connector.get_collection_items_pg(group_id, z["key"],num_pages+1)
                for x in items_dict["items"]:
                    key_list.add(x["key"])
        key_dict[z["key"]] = key_list
    return key_dict
```

```python
#collections_dict #<group_id>: [collection ids]
#{group_id: {collection_id: []}}
files_dict={}
for group_ids in collections_dict:
    if group_ids not in files_dict:
        files_dict[group_ids] = {}
    files_dict[group_ids].update(get_items(group_ids))        
```
